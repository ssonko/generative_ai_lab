```python
import os
from dotenv import load_dotenv

from langchain_openai import ChatOpenAI
# from langchain_ollama import ChatOllama
from langchain_core.prompts import PromptTemplate

# --- Load API Key ---
load_dotenv(override=True, dotenv_path="../.env")
my_api_key = os.getenv("OPENAI_API_KEY")

# Choose the OpenAI LLM
llm = ChatOpenAI(
    model="gpt-5-nano",
    openai_api_key=my_api_key
)


# # Choose the Ollama LLM
# llm_ollama = ChatOllama(
#     model="gpt-5-nano",
#     openai_api_key=my_api_key
# )

# Create a prompt template
prompt = PromptTemplate(
    template="You are an AI assistant. Respond to this concept in one line: {concept}",
    input_variables=["concept"],
)

# Build the chain (modern runnable syntax)
chain = prompt | llm

# chain_ollama = prompt | llm_ollama

# Run the chain
response = chain.invoke(
    {"concept": "What is Retrieval Augmented Generation?"}
)

print(response.content)
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI

classifier_prompt = PromptTemplate(
    template="""
        You are a classifier agent in a library system.

        Determine whether the question involves:
        1. Library FAQs (hours, rules, membership)
        2. Book availability or checkout

        Return ONLY valid JSON in this exact format:
        {{
        "faq_flag": true,
        "checkout_flag": false
        }}

        Question: {question}
        """,
    input_variables=["question"],
)

classifier_llm = ChatOpenAI(
    model="gpt-5-nano",
    temperature=0,
)

chain = classifier_prompt | classifier_llm

response = chain.invoke(
    {"question": "When does library open?"}
)

print(response.content)
```
